```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')

star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

labels = {
    'Have you seen any of the 6 films in the Star Wars franchise?': 'seen',
    'Do you consider yourself to be a fan of the Star Wars film franchise?': 'fan'
}

star_wars = star_wars.rename(columns=labels)
cols = ['seen', 'fan']

star_wars[cols[0]] = star_wars[cols[0]].map(yes_no)
star_wars[cols[1]] = star_wars[cols[1]].map(yes_no)
star_wars.head()
```

```python
import numpy as np

seens = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seens)
    
labels = dict()
for i, c in enumerate(star_wars.columns[3:9]):
    labels[c] = 'seen_' + str(i+1)
    
star_wars = star_wars.rename(columns=labels)
    
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

labels = dict()

for i, c in enumerate(star_wars.columns[9:15]):
    labels[c] = 'ranking_' + str(i+1)

star_wars = star_wars.rename(columns=labels)
star_wars.head()
```

```python
%matplotlib inline
ranking_means = star_wars[star_wars.columns[9:15]].mean()
ranking_means.plot.bar()
```

The trilogy of Anakin Skywalker is the most rated, with The Revenge of the Sith topping the charts. This may be due to the a predominance of a younger age group that grew up after the first trilogy. To assess this further we will check what is the mean age of this group.

```python
seen_group = star_wars[star_wars['seen'] == True]
ranking_means = seen_group[seen_group.columns[9:15]].mean()
ranking_means.plot.bar()
```

```python
fan_group = seen_group[seen_group['fan'] == True]
ranking_means = fan_group[fan_group.columns[9:15]].mean()
ranking_means.plot.bar()
```

Assuming the movie chronology is wrong, we can assess that who have seen SW prefer the original movies. This is even clearer when only the fan group is taken into account.

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

It's normal that the 2nd and 3rd original movies were the most seen. No one knew what the 1st movie was about, but as it gained momentum, the 2nd and 3rd were more watched.
Regarding the 2nd trilogy, the 1st was highly anticipated as the return of a famous franchise, but it was a disappointment. That reflected on a lowering of the 2 sequels.
