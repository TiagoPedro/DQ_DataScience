```python
import pandas as pd

movies = pd.read_csv('fandango_score_comparison.csv')
movies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(movies['Metacritic_norm_round'])
plt.xlim(0.0, 5.0)
plt.show()

plt.hist(movies['Fandango_Stars'])
plt.xlim(0.0, 5.0)
plt.show()
```

The Fandango scores are much less distributed and are clustered on the high-end of the scale.

```python
import numpy as np

f_mean = np.mean(movies['Fandango_Stars'])
m_mean = np.mean(movies['Metacritic_norm_round'])

f_median = np.median(movies['Fandango_Stars'])
m_median = np.median(movies['Metacritic_norm_round'])

f_std = np.std(movies['Fandango_Stars'])
m_std = np.std(movies['Metacritic_norm_round'])

print(f_mean)
print(m_mean)
print(f_median)
print(m_median)
print(f_std)
print(m_std)
```

The standard deviation for Fandango is much lower than the Metacritic one because its results are clustered on the [3,5] interval, while the Metacritic results are spread through the [1,5] interval.

```python
plt.scatter(movies['Fandango_Stars'], movies['Metacritic_norm_round'])
```

```python
movies['fm_diff'] = movies['Fandango_Stars'] - movies['Metacritic_norm_round']
movies['fm_diff'] = np.absolute(movies['fm_diff'])
movies.sort_values(by='fm_diff', ascending=False).head()
```

```python
from scipy.stats import pearsonr

r_value, p_value = pearsonr(movies['Metacritic_norm_round'], movies['Fandango_Stars'])
r_value
```

The low r-value means that there is an inconsistency between Metacritic's and Fandango's reviews. As we know the methodology of Fandango is shady, we can hypothesize that it is biased.

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, stderr_slope = linregress(movies['Metacritic_norm_round'], movies['Fandango_Stars'])

pred_3 = 3 * slope + intercept
pred_3
```

```python
pred_1 = 1 * slope + intercept
pred_5 = 5 * slope + intercept

plt.scatter(movies['Metacritic_norm_round'], movies['Fandango_Stars'])
x = [1.0, 5.0]
y = [pred_1, pred_5]
plt.plot(x, y)
plt.show()
```
